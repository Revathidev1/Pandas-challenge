### Categories:
* PCs, Laptops, Servers & Harddrives
* iPhone, iPad, iPod, Apple Watch
* Cables & Adaptors
* Parts and Repair
* Accessories & Memory

```python
from collections import Counter
types = list(products.type.unique())
type_dict = {}
for t in types:
  type_dict[t] = list(products.loc[products.type == t, 'desc'])
```

```python
type_words = {}
for td in type_dict.keys():
  string = ''
  for value in type_dict[td]:
    string += ' '+ value
    list = string.split()
    new_list = []
    for l in list:
      l = l.replace('.','').replace('+', '/').replace('/','')
      new_list.append(l)
    for item in new_list:
      if item in ['', ' ', 'your', 'and', 'with', 'for', 'of', 'to']:
        new_list.remove(item)
  # print(new_list)
    counter = Counter(new_list)
    most_occur = counter.most_common(5)
    words = []
    for i in range(len(most_occur)):
      words.append(most_occur[i][0])
    type_words[td] = words
```

```python
keys = ['PCs, Laptops, Servers & Harddrives', 'iPhone, iPad, iPod, Apple Watch', 'Cables & Adaptors', 'Parts, Repair, & Warranties', 'Accessories & Memory']
categories = {key: [] for key in keys}
```

```python
for tw in type_words:
  for item in type_words[tw]:
    if item in ['support', 'stand', 'keyboard', 'mouse', 'headphone', 'RAM', 'case', 'Case', 'headset', 'Wacom', 'tips', 'card', 'backpack', 'microphone', 'Remote', 'player', 'DVD', 'two-wheeled', 'speaker', 'pen', 'armband', 'external']:
      categories['Accessories & Memory'].append(tw)
      type_words[tw].clear()
    if item in ['Lightning', 'cable', 'Cable', 'Thunderbolt', 'Ethernet', 'Charger', 'charger', 'connector', 'adaptor','adapter', 'SATA', 'outlets', 'HomeKit']:
      categories['Cables & Adaptors'].append(tw)
      type_words[tw].clear()
    if item in ['parts', ' repair', 'diagnosis', 'extended', 'Protector', 'Cleaner', 'Replacement', 'replacement']:
      categories['Parts, Repair, & Warranties'].append(tw)
      type_words[tw].clear()
    if item in ['Synology', 'SSD', 'MacBook', 'PC', 'iMac', 'Computer', 'Mac', 'Expansion', 'Core']:
      categories['PCs, Laptops, Servers & Harddrives'].append(tw)
      type_words[tw].clear()  
    if item in ['iPhone', 'iPad', 'Watch', 'iPod']:
      categories['iPhone, iPad, iPod, Apple Watch'].append(tw)
      type_words[tw].clear()
```

```python
def cat_app(type):
  for cat in categories:
    if type in categories[cat]:
      return cat
  
products['categories'] = products.type.apply(cat_app)
```
